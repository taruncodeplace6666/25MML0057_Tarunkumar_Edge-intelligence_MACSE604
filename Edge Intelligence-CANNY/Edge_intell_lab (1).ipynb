```python
# Import necessary libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files
import os
```

```python
# Path to the video file in Colab environment
video_path = '/content/file_example_MP4_480_1_5MG.mp4'

# Open the video file
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Error: Could not open video {video_path}")
else:
    # Read the first frame
    ret, frame = cap.read()

    if ret:
        print("Frame successfully extracted.")
        # Display the frame
        cv2_imshow(frame)
        # Store the frame for the next steps
        color_frame = frame
    else:
        print("Error: Could not read frame.")


cap.release()
```

```python

if 'color_frame' in locals():
    print(f"Color frame shape: {color_frame.shape}")
    pixel_coords = (100, 100)
    pixel_value_bgr = color_frame[pixel_coords[0], pixel_coords[1]]
    print(f"Pixel value at {pixel_coords} (BGR): {pixel_value_bgr}")
else:
    print("Color frame not found. Please run Cell 2 first.")
```

```python

if 'color_frame' in locals():

    gray_frame = cv2.cvtColor(color_frame, cv2.COLOR_BGR2GRAY)

    print("Frame successfully converted to grayscale.")


    print(f"Grayscale frame shape: {gray_frame.shape}")


    pixel_coords = (100, 100)
    pixel_value_gray = gray_frame[pixel_coords[0], pixel_coords[1]]
    print(f"Pixel value at {pixel_coords} (Grayscale): {pixel_value_gray}")

    # Display the grayscale frame
    cv2_imshow(gray_frame)
else:
    print("Color frame not found. Please run Cell 2 first.")
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
import io
from PIL import Image

uploaded = files.upload()
video_path = list(uploaded.keys())[0]

# Open the video file
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video file.")
else:
    print(f"Video '{video_path}' successfully loaded.")
```

```python
frame_number = 50
cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number) # Set the frame position

ret, frame = cap.read() # Read the frame

if ret:
    print(f"Frame {frame_number} extracted successfully.")
    # Release the video capture object
    cap.release()
else:
    print(f"Error: Could not read frame {frame_number}.")
```

```python
if ret:
    # Convert the frame from BGR (OpenCV default) to RGB (Matplotlib default)
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Display the frame
    plt.imshow(frame_rgb)
    plt.title(f"Frame {frame_number} (Color)")
    plt.axis('off')
    plt.show()

    # Print the pixel values in matrix form
    print(f"Pixel values of the color frame (first few rows/columns):\n{frame[:5, :5, :]}")
    print(f"Frame shape: {frame.shape}")
else:
    print("No frame to display or print.")
```

```python
if ret:
    # Convert the BGR frame to grayscale
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the grayscale frame
    plt.imshow(frame_gray, cmap='gray', vmin=0, vmax=255)
    plt.title(f"Frame {frame_number} (Grayscale)")
    plt.axis('off')
    plt.show()

    # Print the pixel values in matrix form (grayscale images are 2D arrays)
    print(f"Pixel values of the grayscale frame (first few rows/columns):\n{frame_gray[:5, :5]}")
    print(f"Grayscale frame shape: {frame_gray.shape}")
else:
    print("No frame to convert or print.")
```

```python
import cv2
import numpy as np
from google.colab import files
import os



# Assuming you uploaded a file named 'input_video.mp4'
input_video_path = '/content/file_example_MP4_480_1_5MG (1).mp4'
output_video_path = 'output_edges.mp4'



cap = cv2.VideoCapture(input_video_path)

# Check if the video opened successfully
if not cap.isOpened():
    print("Error opening video stream or file")
    exit()

# 2. Get video properties (width, height, FPS) for the output video
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)


# 'MP4V' is a commonly supported codec for .mp4 files
fourcc = cv2.VideoWriter_fourcc(*'MP4V')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height), isColor=False) # Canny output is grayscale

# 4. Process the video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        # Apply Canny edge detection
        # Canny operates on grayscale, so convert if needed, although cv2.Canny can work on BGR
        edges = cv2.Canny(frame, 100, 200)

        # Write the processed frame to the output file
        # The output writer was created as grayscale (isColor=False), so we write the single-channel 'edges'
        out.write(edges)

    else:
        break

# 5. Release everything when the job is finished
cap.release()
out.release()
print(f"Processing complete. Output saved as {output_video_path}")


try:
    files.download(output_video_path)
except Exception as e:
    print(f"Could not initiate download: {e}")

```

```python
from IPython.display import HTML
from base64 import b64encode
import os

# 1. Convert the video to H.264 using ffmpeg
compressed_path = "output_compressed.mp4"
if os.path.exists(output_video_path):
    os.system(f"ffmpeg -i {output_video_path} -vcodec libx264 {compressed_path} -y -loglevel quiet")

# 2. Display the video in the output cell
mp4 = open(compressed_path,'rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()

HTML(f"""
<video width={frame_width // 2} controls>
      <source src="{data_url}" type="video/mp4">
</video>
""")
```
