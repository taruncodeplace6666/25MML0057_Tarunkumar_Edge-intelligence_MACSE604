```python
from nltk.corpus import names
names.words()
print(len(names.words()))
```

```python
def gender_features(word):
    return {'last_letter': word[-1]}  # Use square brackets for indexing
```

```python
labeled_names = ([(name, 'male') for name in  names.words('male.txt')] + [(name, 'female') for name  in names.words('female.txt')])
```

```python
import nltk
import random

# 1. Assuming labeled_names is already loaded from nltk.corpus.names
random.shuffle(labeled_names)

# 2. Extract features into a list of tuples (features, label)
# Ensure gender_features is defined (e.g., returning {'last_letter': n[-1]})
featuresets = [(gender_features(n), gender) for (n, gender) in labeled_names]

# 3. Correctly split the data (using the same variable name)
# Typically, use the first 500 for testing and the rest for training
test_set = featuresets[:2000]
train_set = featuresets[5000:]

# 4. Train the classifier
classifier = nltk.NaiveBayesClassifier.train(train_set)

# 5. Classify and test accuracy
print(classifier.classify(gender_features('David')))
print(classifier.classify(gender_features('Yann Lann')))
print(f"Accuracy: {nltk.classify.accuracy(classifier, test_set):.2%}")
```
