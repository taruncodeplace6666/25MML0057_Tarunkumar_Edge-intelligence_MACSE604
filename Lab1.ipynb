```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np

#DATASET LINK:https://www.kaggle.com/datasets/dschettler8845/labelme-12-50k
# Define the exact path to your image file in the Colab session
image_file_path = r'/content/039000.jpg'

if os.path.exists(image_file_path):
    img_bgr = cv2.imread(image_file_path)

    if img_bgr is not None:
        height, width, channels = img_bgr.shape
        print(f"Original shape (H, W, C): {img_bgr.shape}\n")

        # --- 1. Resizing the image (changing dimensions) ---
        TARGET_SIZE = (100, 100) # OpenCV expects (width, height) for this function
        resized_img = cv2.resize(img_bgr, TARGET_SIZE, interpolation=cv2.INTER_AREA)

        print(f"Resized shape (H, W, C): {resized_img.shape}")

        # --- 2. Reshaping the image (flattening the array) ---
        # This turns the 3D array (Height, Width, Channels) into a 1D array (Total Pixels)
        # Useful for input into simple machine learning models (like simple Logistic Regression)
        flattened_img = resized_img.reshape(-1)

        print(f"Flattened shape: {flattened_img.shape}")
        print(f"Total number of elements in flattened array: {flattened_img.size}")

        # --- Display the Resized Image ---
        img_rgb = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)

        plt.figure(figsize=(4, 4))
        plt.imshow(img_rgb)
        plt.title(f"Resized Image: {resized_img.shape}")
        plt.axis('off')
        plt.show()

    else:
        print("Error: OpenCV could not read the image data.")
else:
    print("ERROR: The file path provided does not exist in the Colab environment.")
```

```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np

# Define the exact path to your image file in the Colab session
image_file_path = r'/content/039000.jpg'

print(f"Attempting to load and analyze image: {image_file_path}\n")

if os.path.exists(image_file_path):
    # Load the image using OpenCV (loads as BGR format NumPy array)
    img_bgr = cv2.imread(image_file_path)

    if img_bgr is not None:
        # --- Basic Information ---
        print("--- Image Information ---")
        print(f"Image loaded successfully.")
        print(f"NumPy Array Type: {type(img_bgr)}")

        # --- Shape and Dimensions ---
        height, width, channels = img_bgr.shape
        print(f"Shape (Height, Width, Channels): {img_bgr.shape}")
        print(f"Height: {height} pixels")
        print(f"Width: {width} pixels")
        print(f"Channels (Color Depth): {channels} (BGR)")

        # --- Data Type and Size ---
        print(f"Pixel data type (Dtype): {img_bgr.dtype}") # e.g., uint8 (unsigned integer 8-bit)
        print(f"Minimum pixel value: {img_bgr.min()}")
        print(f"Maximum pixel value: {img_bgr.max()}")
        print(f"Total number of pixels: {img_bgr.size // channels}") # total elements divided by channels

        # --- Display the image ---
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

        plt.figure(figsize=(8, 8))
        plt.imshow(img_rgb)
        plt.title(f"Loaded Image: {height}x{width}")
        plt.axis('off')
        plt.show()
    else:
        print("Error: OpenCV could not read the image data. The file might be corrupted.")
else:
    print("ERROR: The file path provided does not exist in the Colab environment.")
    print("Please ensure you have uploaded the file to the current session.")
```
